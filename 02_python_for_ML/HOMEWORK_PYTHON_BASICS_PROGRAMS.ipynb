## Program 9: Grade from Marks

**Problem**: Read marks (0–100) and print grade.

**Grading**:
- 90–100: A
- 80–89: B
- 70–79: C
- 60–69: D
- <60: F

**Input**:
- One integer `marks`

**Output**:
- Print the grade letter

**Sample Input**:
```
83
```
**Sample Output**:
```
B
```


````python
marks = int(input("Enter marks: "))

match marks:
    case _ if 90 <= marks <= 100:
        print("A")
    case _ if 80 <= marks <= 89:
        print("B")
    case _ if 70 <= marks <= 79:
        print("C")
    case _ if 60 <= marks <= 69:
        print("D")
    case _ if 0 <= marks < 60:
        print("F")
    case _:
        print("Invalid Input")
````

## Program 13: Fibonacci (N Terms)

**Problem**: Read `N` and print first `N` Fibonacci numbers.

**Input**:
- One integer `N`

**Output**:
- Print Fibonacci numbers separated by space

**Sample Input**:
```
7
```
**Sample Output**:
```
0 1 1 2 3 5 8
```


````python
number_of_terms = int(input("Enter number of Fibonacci terms: ").strip())

fibonacci_numbers = []

if number_of_terms > 0:
    fibonacci_numbers.append(0)

if number_of_terms > 1:
    fibonacci_numbers.append(1)

for i in range(2, number_of_terms):
    next_value = fibonacci_numbers[i - 1] + fibonacci_numbers[i - 2]
    fibonacci_numbers.append(next_value)

print(" ".join(map(str, fibonacci_numbers)))
````

## Program 14: Prime Check

**Problem**: Read an integer `n` and print `Prime` or `Not Prime`.

**Input**:
- One integer `n`

**Output**:
- `Prime` or `Not Prime`

**Sample Input**:
```
29
```
**Sample Output**:
```
Prime
```


````python
number = int(input("Enter a number to check if it is prime: ").strip())

is_prime = True

if number < 2:
    is_prime = False
else:
    for divisor in range(2, int(number ** 0.5) + 1):
        if number % divisor == 0:
            is_prime = False
            break

if is_prime:
    print("Prime")
else:
    print("Not Prime")
````

## Program 28: Second Largest (Distinct)

**Problem**: Read `N` and a list of integers. Print the second largest **distinct** number.

**If second largest does not exist**, print: `Not Possible`.

**Input**:
- Integer `N`
- Next line: `N` integers

**Output**:
- Second largest distinct value or `Not Possible`

**Sample Input**:
```
5
5 1 5 3 2
```
**Sample Output**:
```
3
```


````python
count_of_numbers = int(input("Enter the number of integers: ").strip())
numbers_list = list(map(int, input("Enter the integers (space-separated): ").split()))

unique_numbers = sorted(set(numbers_list[:count_of_numbers]), reverse=True)

if len(unique_numbers) < 2:
    print("Not Possible")
else:
    print(unique_numbers[1])
````

## Program 29: Rotate List Right by K

**Problem**: Read `N`, list of `N` integers, and `K`. Rotate the list to the right by `K` positions.

**Input**:
- Integer `N`
- Next line: `N` integers
- Integer `K`

**Output**:
- Rotated list

**Sample Input**:
```
5
1 2 3 4 5
2
```
**Sample Output**:
```
4 5 1 2 3
```


````python
count_of_numbers = int(input("Enter the number of integers: ").strip())
numbers_list = input("Enter the integers (space-separated): ").split()
rotation_count = int(input("Enter the rotation count K: ").strip())

if count_of_numbers == 0:
    print()
else:
    k = rotation_count % count_of_numbers
    rotated_list = numbers_list[-k:] + numbers_list[:-k]
    print(" ".join(rotated_list))
````
